```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Sample - Superstore.csv", encoding='latin-1')
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
df.info()
```

```python
df["Month"] = df["Order Date"].dt.month
df["Year"] = df["Order Date"].dt.year
df
```

```python
df = df.sort_values(by="Order Date")
df
```

```python
profitable_region = df.groupby("Region")["Profit"].sum()
sorted_profitable_region = profitable_region.sort_values(ascending=False)
sorted_profitable_region
```

```python
Products_by_sales = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False)
Products_by_sales
```

```python
top_profit_products = df.groupby("Product Name")["Profit"].sum().sort_values(ascending=False).head(5)
print(top_profit_products)
```

```python
loss_data = df[df["Profit"] < 0]
```

```python
loss_by_category = loss_data.groupby("Category")["Profit"].sum()
print(loss_by_category)
```

```python
loss_by_region = loss_data.groupby("Region")["Profit"].sum().sort_values()
print(loss_by_region)
```

```python
loss_by_category.plot(kind='barh', title="Losses by Category", color='red')
plt.xlabel("Total Loss (Â£)")
plt.tight_layout()
plt.show()
```

```python
monthly_sales = df.groupby("Month")["Sales"].sum()
monthly_sales
```

```python
monthly_sales.plot(kind='line', title="Monthly Sales", color='green',marker = 'o')
plt.xlabel("Month")
plt.ylabel("Total Sales (Â£)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
